```python
"""
Atividade Prática VIII - Human Activity Recognition (HAR)
Comparativo de Classificação: KNN, SVM e MLP
Com e Sem LDA
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import cross_val_score, train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score, 
                             f1_score, confusion_matrix, classification_report)
import warnings
warnings.filterwarnings('ignore')

# ============================================================================
# 1. CARREGAMENTO E PREPARAÇÃO DOS DADOS
# ============================================================================

def load_har_dataset(base_path):
    """
    Carrega o dataset HAR a partir dos arquivos de treino e teste
    """
    print("Carregando dados...")
    
    # Carregar features
    features = pd.read_csv(f'{base_path}/features.txt', 
                          sep='\s+', header=None, names=['id', 'feature'])
    feature_names = features['feature'].tolist()
    
    # Tornar nomes de features únicos (alguns são duplicados no dataset original)
    feature_counts = {}
    unique_feature_names = []
    for name in feature_names:
        if name in feature_counts:
            feature_counts[name] += 1
            unique_feature_names.append(f"{name}_{feature_counts[name]}")
        else:
            feature_counts[name] = 0
            unique_feature_names.append(name)
    
    print(f"Features duplicadas corrigidas: {len(feature_names)} -> {len(set(unique_feature_names))} únicas")
    
    # Carregar labels de atividades
    activity_labels = pd.read_csv(f'{base_path}/activity_labels.txt',
                                 sep='\s+', header=None, 
                                 names=['id', 'activity'])
    
    # Carregar dados de treino
    X_train = pd.read_csv(f'{base_path}/train/X_train.txt',
                         sep='\s+', header=None, names=unique_feature_names)
    y_train = pd.read_csv(f'{base_path}/train/y_train.txt',
                         header=None, names=['activity'])
    
    # Carregar dados de teste
    X_test = pd.read_csv(f'{base_path}/test/X_test.txt',
                        sep='\s+', header=None, names=unique_feature_names)
    y_test = pd.read_csv(f'{base_path}/test/y_test.txt',
                        header=None, names=['activity'])
    
    # Combinar treino e teste para reorganizar depois
    X = pd.concat([X_train, X_test], axis=0).reset_index(drop=True)
    y = pd.concat([y_train, y_test], axis=0).reset_index(drop=True)
    
    print(f"Dataset carregado: {X.shape[0]} amostras, {X.shape[1]} features")
    print(f"Distribuição das classes:\n{y['activity'].value_counts().sort_index()}")
    
    return X, y, activity_labels

# ============================================================================
# 2. NORMALIZAÇÃO DOS DADOS
# ============================================================================

def normalize_data(X):
    """
    Normaliza os dados usando StandardScaler
    """
    scaler = StandardScaler()
    X_normalized = scaler.fit_transform(X)
    print("Dados normalizados com StandardScaler")
    return X_normalized, scaler

# ============================================================================
# 3. APLICAÇÃO DO LDA
# ============================================================================

def apply_lda(X, y, n_components=5):
    """
    Aplica Linear Discriminant Analysis para redução de dimensionalidade
    """
    print(f"\nAplicando LDA com {n_components} componentes...")
    lda = LinearDiscriminantAnalysis(n_components=n_components)
    X_lda = lda.fit_transform(X, y.values.ravel())
    
    variance_explained = np.sum(lda.explained_variance_ratio_)
    print(f"Variância explicada pelo LDA: {variance_explained:.4f}")
    
    return X_lda, lda

# ============================================================================
# 4. TREINAMENTO E AVALIAÇÃO DOS MODELOS
# ============================================================================

def evaluate_model(model, X_train, X_test, y_train, y_test, model_name):
    """
    Treina e avalia um modelo, retornando métricas
    """
    # Treinar modelo
    model.fit(X_train, y_train.values.ravel())
    
    # Predições
    y_pred = model.predict(X_test)
    
    # Calcular métricas
    metrics = {
        'model': model_name,
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred, average='weighted'),
        'recall': recall_score(y_test, y_pred, average='weighted'),
        'f1_score': f1_score(y_test, y_pred, average='weighted')
    }
    
    # Matriz de confusão
    cm = confusion_matrix(y_test, y_pred)
    
    return metrics, cm, y_pred

def cross_validate_model(model, X, y, model_name, cv=10):
    """
    Realiza validação cruzada e retorna as métricas
    """
    print(f"\nValidação Cruzada - {model_name}...")
    
    cv_accuracy = cross_val_score(model, X, y.values.ravel(), 
                                  cv=cv, scoring='accuracy')
    cv_precision = cross_val_score(model, X, y.values.ravel(), 
                                   cv=cv, scoring='precision_weighted')
    cv_recall = cross_val_score(model, X, y.values.ravel(), 
                                cv=cv, scoring='recall_weighted')
    cv_f1 = cross_val_score(model, X, y.values.ravel(), 
                           cv=cv, scoring='f1_weighted')
    
    cv_metrics = {
        'model': model_name,
        'cv_accuracy_mean': cv_accuracy.mean(),
        'cv_accuracy_std': cv_accuracy.std(),
        'cv_precision_mean': cv_precision.mean(),
        'cv_recall_mean': cv_recall.mean(),
        'cv_f1_mean': cv_f1.mean()
    }
    
    return cv_metrics

# ============================================================================
# 5. CONFIGURAÇÃO E OTIMIZAÇÃO DOS MODELOS
# ============================================================================

def train_knn(X_train, X_test, y_train, y_test, scenario):
    """
    Treina KNN com otimização do parâmetro K
    """
    print(f"\n{'='*60}")
    print(f"TREINANDO KNN - {scenario}")
    print(f"{'='*60}")
    
    # Testar diferentes valores de K
    k_values = [3, 5, 7, 9, 11, 15, 19]
    best_score = 0
    best_k = 3
    
    print("Testando valores de K:", k_values)
    for k in k_values:
        knn = KNeighborsClassifier(n_neighbors=k, n_jobs=-1)
        score = cross_val_score(knn, X_train, y_train.values.ravel(), 
                               cv=5, scoring='accuracy').mean()
        print(f"K={k}: Acurácia CV = {score:.4f}")
        if score > best_score:
            best_score = score
            best_k = k
    
    print(f"\nMelhor K: {best_k} (Acurácia CV: {best_score:.4f})")
    
    # Treinar modelo final
    knn_final = KNeighborsClassifier(n_neighbors=best_k, n_jobs=-1)
    metrics, cm, y_pred = evaluate_model(knn_final, X_train, X_test, 
                                        y_train, y_test, f'KNN (K={best_k})')
    
    return knn_final, metrics, cm

def train_svm(X_train, X_test, y_train, y_test, scenario):
    """
    Treina SVM com kernel RBF e otimização do parâmetro C
    """
    print(f"\n{'='*60}")
    print(f"TREINANDO SVM - {scenario}")
    print(f"{'='*60}")
    
    # Grid Search para otimizar C e gamma
    param_grid = {
        'C': [0.1, 1, 10, 100],
        'gamma': ['scale', 'auto', 0.001, 0.01]
    }
    
    svm = SVC(kernel='rbf', random_state=42)
    
    print("Executando Grid Search para otimizar C e gamma...")
    grid_search = GridSearchCV(svm, param_grid, cv=5, 
                              scoring='accuracy', n_jobs=-1, verbose=1)
    grid_search.fit(X_train, y_train.values.ravel())
    
    print(f"\nMelhores parâmetros: {grid_search.best_params_}")
    print(f"Melhor acurácia CV: {grid_search.best_score_:.4f}")
    
    # Avaliar modelo final
    svm_final = grid_search.best_estimator_
    metrics, cm, y_pred = evaluate_model(svm_final, X_train, X_test, 
                                        y_train, y_test, 
                                        f"SVM (C={grid_search.best_params_['C']}, "
                                        f"gamma={grid_search.best_params_['gamma']})")
    
    return svm_final, metrics, cm

def train_mlp(X_train, X_test, y_train, y_test, scenario):
    """
    Treina MLP com otimização de arquitetura e regularização
    """
    print(f"\n{'='*60}")
    print(f"TREINANDO MLP - {scenario}")
    print(f"{'='*60}")
    
    # Regra heurística: neurônios = (n_features + n_classes) / 2
    n_features = X_train.shape[1]
    n_classes = len(np.unique(y_train))
    n_neurons = (n_features + n_classes) // 2
    
    print(f"Features: {n_features}, Classes: {n_classes}")
    print(f"Neurônios por camada (heurística): {n_neurons}")
    
    # Testar diferentes arquiteturas e alpha
    architectures = [
        (n_neurons,),
        (n_neurons, n_neurons//2),
        (n_neurons*2, n_neurons),
    ]
    
    alphas = [0.0001, 0.001, 0.01]
    
    best_score = 0
    best_config = None
    
    print("\nTestando configurações...")
    for arch in architectures:
        for alpha in alphas:
            mlp = MLPClassifier(hidden_layer_sizes=arch, 
                               alpha=alpha,
                               max_iter=500,
                               random_state=42,
                               early_stopping=True,
                               validation_fraction=0.1)
            
            score = cross_val_score(mlp, X_train, y_train.values.ravel(),
                                   cv=3, scoring='accuracy').mean()
            
            print(f"Arquitetura: {arch}, Alpha: {alpha} -> Acurácia: {score:.4f}")
            
            if score > best_score:
                best_score = score
                best_config = (arch, alpha)
    
    print(f"\nMelhor configuração: Arquitetura={best_config[0]}, "
          f"Alpha={best_config[1]} (Acurácia CV: {best_score:.4f})")
    
    # Treinar modelo final
    mlp_final = MLPClassifier(hidden_layer_sizes=best_config[0],
                             alpha=best_config[1],
                             max_iter=1000,
                             random_state=42,
                             early_stopping=True,
                             validation_fraction=0.1)
    
    metrics, cm, y_pred = evaluate_model(mlp_final, X_train, X_test,
                                        y_train, y_test,
                                        f"MLP {best_config[0]} (α={best_config[1]})")
    
    return mlp_final, metrics, cm

# ============================================================================
# 6. VISUALIZAÇÃO DOS RESULTADOS
# ============================================================================

def plot_confusion_matrix(cm, title, activity_labels):
    """
    Plota matriz de confusão
    """
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=activity_labels['activity'],
                yticklabels=activity_labels['activity'])
    plt.title(title)
    plt.ylabel('Verdadeiro')
    plt.xlabel('Predito')
    plt.tight_layout()
    plt.show()

def plot_comparison(results_df, scenario):
    """
    Plota comparação de métricas entre modelos
    """
    metrics = ['accuracy', 'precision', 'recall', 'f1_score']
    
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    fig.suptitle(f'Comparação de Modelos - {scenario}', fontsize=16)
    
    for idx, metric in enumerate(metrics):
        ax = axes[idx // 2, idx % 2]
        results_df.plot(x='model', y=metric, kind='bar', ax=ax, legend=False)
        ax.set_title(metric.upper())
        ax.set_ylabel('Score')
        ax.set_xlabel('')
        ax.set_ylim([0, 1])
        ax.grid(axis='y', alpha=0.3)
        
        # Adicionar valores nas barras
        for container in ax.containers:
            ax.bar_label(container, fmt='%.3f')
    
    plt.tight_layout()
    plt.show()

# ============================================================================
# 7. PIPELINE PRINCIPAL
# ============================================================================

def run_experiments(base_path):
    """
    Executa todos os experimentos
    """
    # Carregar dados
    X, y, activity_labels = load_har_dataset(base_path)
    
    # Normalizar dados
    X_normalized, scaler = normalize_data(X)
    
    # Dividir em treino (70%) e teste (30%)
    X_train, X_test, y_train, y_test = train_test_split(
        X_normalized, y, test_size=0.3, random_state=42, stratify=y
    )
    
    print(f"\nDivisão dos dados:")
    print(f"Treino: {X_train.shape[0]} amostras")
    print(f"Teste: {X_test.shape[0]} amostras")
    
    # ========================================================================
    # CENÁRIO 1: SEM LDA
    # ========================================================================
    print("\n" + "="*70)
    print("CENÁRIO 1: CLASSIFICAÇÃO SEM LDA")
    print("="*70)
    
    results_no_lda = []
    
    # KNN
    knn_model, knn_metrics, knn_cm = train_knn(X_train, X_test, y_train, 
                                               y_test, "Sem LDA")
    results_no_lda.append(knn_metrics)
    plot_confusion_matrix(knn_cm, f"KNN - Sem LDA", activity_labels)
    
    # SVM
    svm_model, svm_metrics, svm_cm = train_svm(X_train, X_test, y_train, 
                                               y_test, "Sem LDA")
    results_no_lda.append(svm_metrics)
    plot_confusion_matrix(svm_cm, f"SVM - Sem LDA", activity_labels)
    
    # MLP
    mlp_model, mlp_metrics, mlp_cm = train_mlp(X_train, X_test, y_train, 
                                               y_test, "Sem LDA")
    results_no_lda.append(mlp_metrics)
    plot_confusion_matrix(mlp_cm, f"MLP - Sem LDA", activity_labels)
    
    # Resultados Cenário 1
    results_no_lda_df = pd.DataFrame(results_no_lda)
    print("\n" + "="*70)
    print("RESULTADOS FINAIS - SEM LDA")
    print("="*70)
    print(results_no_lda_df.to_string(index=False))
    plot_comparison(results_no_lda_df, "Sem LDA")
    
    # ========================================================================
    # CENÁRIO 2: COM LDA
    # ========================================================================
    print("\n" + "="*70)
    print("CENÁRIO 2: CLASSIFICAÇÃO COM LDA")
    print("="*70)
    
    # Aplicar LDA (máximo de componentes = n_classes - 1 = 5)
    X_train_lda, lda_model = apply_lda(X_train, y_train, n_components=5)
    X_test_lda = lda_model.transform(X_test)
    
    print(f"Dimensão após LDA: {X_train_lda.shape[1]} componentes")
    
    results_lda = []
    
    # KNN com LDA
    knn_lda_model, knn_lda_metrics, knn_lda_cm = train_knn(
        X_train_lda, X_test_lda, y_train, y_test, "Com LDA"
    )
    results_lda.append(knn_lda_metrics)
    plot_confusion_matrix(knn_lda_cm, f"KNN - Com LDA", activity_labels)
    
    # SVM com LDA
    svm_lda_model, svm_lda_metrics, svm_lda_cm = train_svm(
        X_train_lda, X_test_lda, y_train, y_test, "Com LDA"
    )
    results_lda.append(svm_lda_metrics)
    plot_confusion_matrix(svm_lda_cm, f"SVM - Com LDA", activity_labels)
    
    # MLP com LDA
    mlp_lda_model, mlp_lda_metrics, mlp_lda_cm = train_mlp(
        X_train_lda, X_test_lda, y_train, y_test, "Com LDA"
    )
    results_lda.append(mlp_lda_metrics)
    plot_confusion_matrix(mlp_lda_cm, f"MLP - Com LDA", activity_labels)
    
    # Resultados Cenário 2
    results_lda_df = pd.DataFrame(results_lda)
    print("\n" + "="*70)
    print("RESULTADOS FINAIS - COM LDA")
    print("="*70)
    print(results_lda_df.to_string(index=False))
    plot_comparison(results_lda_df, "Com LDA")
    
    # ========================================================================
    # COMPARAÇÃO FINAL ENTRE CENÁRIOS
    # ========================================================================
    print("\n" + "="*70)
    print("COMPARAÇÃO FINAL: SEM LDA vs COM LDA")
    print("="*70)
    
    # Adicionar identificador de cenário
    results_no_lda_df['scenario'] = 'Sem LDA'
    results_lda_df['scenario'] = 'Com LDA'
    
    # Combinar resultados
    all_results = pd.concat([results_no_lda_df, results_lda_df], 
                           ignore_index=True)
    
    # Criar visualização comparativa
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle('Comparação Geral: Sem LDA vs Com LDA', fontsize=16)
    
    metrics = ['accuracy', 'precision', 'recall', 'f1_score']
    
    for idx, metric in enumerate(metrics):
        ax = axes[idx // 2, idx % 2]
        
        # Dados para o gráfico
        data_pivot = all_results.pivot(index='model', 
                                       columns='scenario', 
                                       values=metric)
        
        data_pivot.plot(kind='bar', ax=ax, width=0.8)
        ax.set_title(metric.upper(), fontsize=12)
        ax.set_ylabel('Score')
        ax.set_xlabel('')
        ax.set_ylim([0, 1])
        ax.legend(title='Cenário')
        ax.grid(axis='y', alpha=0.3)
        
        # Rotacionar labels
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
    
    plt.tight_layout()
    plt.show()
    
    # Tabela comparativa final
    print("\nTabela Comparativa Completa:")
    print(all_results[['model', 'scenario', 'accuracy', 'precision', 
                      'recall', 'f1_score']].to_string(index=False))
    
    return all_results

# ============================================================================
# EXECUÇÃO
# ============================================================================

if __name__ == "__main__":
    # Definir o caminho base do dataset
    BASE_PATH = "./uci"
    
    # Executar experimentos
    final_results = run_experiments(BASE_PATH)
    
```
