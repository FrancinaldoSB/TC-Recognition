```python
"""
Atividade Prática VIII - Human Activity Recognition (HAR)
Comparativo de Classificação: KNN, SVM e MLP
Com e Sem LDA
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import cross_val_score, train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score, 
                             f1_score, confusion_matrix)
import warnings
warnings.filterwarnings('ignore')

# ============================================================================
# 1. CARREGAMENTO E PREPARAÇÃO DOS DADOS
# ============================================================================

def load_har_dataset(base_path):
    print("Carregando dados...")
    
    features = pd.read_csv(f'{base_path}/features.txt', sep='\s+', header=None, names=['id', 'feature'])
    feature_names = features['feature'].tolist()
    
    feature_counts = {}
    unique_feature_names = []
    for name in feature_names:
        if name in feature_counts:
            feature_counts[name] += 1
            unique_feature_names.append(f"{name}_{feature_counts[name]}")
        else:
            feature_counts[name] = 0
            unique_feature_names.append(name)
    
    print(f"Features duplicadas corrigidas: {len(feature_names)} -> {len(set(unique_feature_names))} únicas")
    
    activity_labels = pd.read_csv(f'{base_path}/activity_labels.txt', sep='\s+', header=None, names=['id', 'activity'])
    
    X_train = pd.read_csv(f'{base_path}/train/X_train.txt', sep='\s+', header=None, names=unique_feature_names)
    y_train = pd.read_csv(f'{base_path}/train/y_train.txt', header=None, names=['activity'])
    X_test = pd.read_csv(f'{base_path}/test/X_test.txt', sep='\s+', header=None, names=unique_feature_names)
    y_test = pd.read_csv(f'{base_path}/test/y_test.txt', header=None, names=['activity'])
    
    X = pd.concat([X_train, X_test], axis=0).reset_index(drop=True)
    y = pd.concat([y_train, y_test], axis=0).reset_index(drop=True)
    
    print(f"Dataset carregado: {X.shape[0]} amostras, {X.shape[1]} features")
    print(f"Distribuição das classes:\n{y['activity'].value_counts().sort_index()}")
    
    return X, y, activity_labels

# ============================================================================
# 2. NORMALIZAÇÃO DOS DADOS
# ============================================================================

def normalize_data(X):
    scaler = StandardScaler()
    X_normalized = scaler.fit_transform(X)
    print("Dados normalizados com StandardScaler")
    return X_normalized, scaler

# ============================================================================
# 3. APLICAÇÃO DO LDA
# ============================================================================

def apply_lda(X, y, n_components=5):
    print(f"\nAplicando LDA com {n_components} componentes...")
    lda = LinearDiscriminantAnalysis(n_components=n_components)
    X_lda = lda.fit_transform(X, y.values.ravel())
    variance_explained = np.sum(lda.explained_variance_ratio_)
    print(f"Variância explicada pelo LDA: {variance_explained:.4f}")
    return X_lda, lda

# ============================================================================
# 4. AVALIAÇÃO E CROSS-VALIDATION
# ============================================================================

def evaluate_model(model, X_train, X_test, y_train, y_test, model_name):
    model.fit(X_train, y_train.values.ravel())
    y_pred = model.predict(X_test)
    
    metrics = {
        'model': model_name,
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred, average='weighted'),
        'recall': recall_score(y_test, y_pred, average='weighted'),
        'f1_score': f1_score(y_test, y_pred, average='weighted')
    }
    
    cm = confusion_matrix(y_test, y_pred)
    return metrics, cm, y_pred

# 🔹 Função auxiliar nova para calcular média e desvio padrão via CV
def compute_cv_stats(model, X, y, cv=5):
    metrics = {}
    scoring = {
        'accuracy': 'accuracy',
        'precision': 'precision_weighted',
        'recall': 'recall_weighted',
        'f1': 'f1_weighted'
    }
    for name, score_type in scoring.items():
        scores = cross_val_score(model, X, y.values.ravel(), cv=cv, scoring=score_type)
        metrics[f'{name}_mean'] = scores.mean()
        metrics[f'{name}_std'] = scores.std()
    return metrics

# ============================================================================
# 5. TREINAMENTO E OTIMIZAÇÃO DOS MODELOS (COM DESVIO PADRÃO)
# ============================================================================

def train_knn(X_train, X_test, y_train, y_test, scenario):
    print(f"\n{'='*60}\nTREINANDO KNN - {scenario}\n{'='*60}")
    k_values = [3, 5, 7, 9, 11, 15, 19]
    best_score, best_k = 0, 3
    
    for k in k_values:
        knn = KNeighborsClassifier(n_neighbors=k, n_jobs=-1)
        score = cross_val_score(knn, X_train, y_train.values.ravel(), cv=5, scoring='accuracy').mean()
        print(f"K={k}: Acurácia CV = {score:.4f}")
        if score > best_score:
            best_score, best_k = score, k
    
    print(f"\nMelhor K: {best_k} (Acurácia CV: {best_score:.4f})")
    
    knn_final = KNeighborsClassifier(n_neighbors=best_k, n_jobs=-1)
    metrics, cm, y_pred = evaluate_model(knn_final, X_train, X_test, y_train, y_test, f'KNN (K={best_k})')
    cv_stats = compute_cv_stats(knn_final, X_train, y_train)
    metrics.update({
        'accuracy_std': cv_stats['accuracy_std'],
        'precision_std': cv_stats['precision_std'],
        'recall_std': cv_stats['recall_std'],
        'f1_score_std': cv_stats['f1_std']
    })
    return knn_final, metrics, cm

def train_svm(X_train, X_test, y_train, y_test, scenario):
    print(f"\n{'='*60}\nTREINANDO SVM - {scenario}\n{'='*60}")
    param_grid = {'C': [0.1, 1, 10], 'gamma': ['scale', 0.01]}
    svm = SVC(kernel='rbf', random_state=42)
    grid_search = GridSearchCV(svm, param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=0)
    grid_search.fit(X_train, y_train.values.ravel())
    
    best_params = grid_search.best_params_
    print(f"Melhores parâmetros: {best_params}")
    
    svm_final = grid_search.best_estimator_
    metrics, cm, y_pred = evaluate_model(svm_final, X_train, X_test, y_train, y_test,
                                         f"SVM (C={best_params['C']}, γ={best_params['gamma']})")
    cv_stats = compute_cv_stats(svm_final, X_train, y_train)
    metrics.update({
        'accuracy_std': cv_stats['accuracy_std'],
        'precision_std': cv_stats['precision_std'],
        'recall_std': cv_stats['recall_std'],
        'f1_score_std': cv_stats['f1_std']
    })
    return svm_final, metrics, cm

def train_mlp(X_train, X_test, y_train, y_test, scenario):
    print(f"\n{'='*60}\nTREINANDO MLP - {scenario}\n{'='*60}")
    n_features, n_classes = X_train.shape[1], len(np.unique(y_train))
    n_neurons = (n_features + n_classes) // 2
    architectures = [(n_neurons,), (n_neurons, n_neurons//2)]
    alphas = [0.0001, 0.001]
    best_score, best_config = 0, None
    
    for arch in architectures:
        for alpha in alphas:
            mlp = MLPClassifier(hidden_layer_sizes=arch, alpha=alpha, max_iter=500,
                                random_state=42, early_stopping=True, validation_fraction=0.1)
            score = cross_val_score(mlp, X_train, y_train.values.ravel(), cv=3, scoring='accuracy').mean()
            print(f"Arquitetura={arch}, α={alpha} → Acurácia={score:.4f}")
            if score > best_score:
                best_score, best_config = score, (arch, alpha)
    
    print(f"\nMelhor configuração: {best_config}, Acurácia CV={best_score:.4f}")
    mlp_final = MLPClassifier(hidden_layer_sizes=best_config[0], alpha=best_config[1],
                              max_iter=1000, random_state=42, early_stopping=True)
    
    metrics, cm, y_pred = evaluate_model(mlp_final, X_train, X_test, y_train, y_test,
                                         f"MLP {best_config[0]} (α={best_config[1]})")
    cv_stats = compute_cv_stats(mlp_final, X_train, y_train)
    metrics.update({
        'accuracy_std': cv_stats['accuracy_std'],
        'precision_std': cv_stats['precision_std'],
        'recall_std': cv_stats['recall_std'],
        'f1_score_std': cv_stats['f1_std']
    })
    return mlp_final, metrics, cm

# ============================================================================
# 6. VISUALIZAÇÃO
# ============================================================================

def plot_confusion_matrix(cm, title, activity_labels):
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=activity_labels['activity'],
                yticklabels=activity_labels['activity'])
    plt.title(title)
    plt.ylabel('Verdadeiro')
    plt.xlabel('Predito')
    plt.tight_layout()
    plt.show()

def plot_comparison(results_df, scenario):
    metrics = ['accuracy', 'precision', 'recall', 'f1_score']
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    fig.suptitle(f'Comparação de Modelos - {scenario}', fontsize=16)
    for idx, metric in enumerate(metrics):
        ax = axes[idx // 2, idx % 2]
        results_df.plot(x='model', y=metric, kind='bar', ax=ax, legend=False)
        ax.set_title(metric.upper())
        ax.set_ylim([0, 1])
        for container in ax.containers:
            ax.bar_label(container, fmt='%.3f')
    plt.tight_layout()
    plt.show()

# ============================================================================
# 7. PIPELINE PRINCIPAL
# ============================================================================

def run_experiments(base_path):
    X, y, activity_labels = load_har_dataset(base_path)
    X_normalized, scaler = normalize_data(X)
    X_train, X_test, y_train, y_test = train_test_split(
        X_normalized, y, test_size=0.3, random_state=42, stratify=y
    )

    print(f"\nTreino: {X_train.shape[0]} | Teste: {X_test.shape[0]}")
    
    # ==================== SEM LDA ====================
    print("\n" + "="*70 + "\nCENÁRIO 1: SEM LDA\n" + "="*70)
    results_no_lda = []
    for trainer in [train_knn, train_svm, train_mlp]:
        model, metrics, cm = trainer(X_train, X_test, y_train, y_test, "Sem LDA")
        results_no_lda.append(metrics)
        plot_confusion_matrix(cm, metrics['model'], activity_labels)
    df_no_lda = pd.DataFrame(results_no_lda)
    print(df_no_lda[['model', 'accuracy', 'accuracy_std', 'precision', 'precision_std',
                     'recall', 'recall_std', 'f1_score', 'f1_score_std']].to_string(index=False))
    plot_comparison(df_no_lda, "Sem LDA")

    # ==================== COM LDA ====================
    print("\n" + "="*70 + "\nCENÁRIO 2: COM LDA\n" + "="*70)
    X_train_lda, lda_model = apply_lda(X_train, y_train, n_components=5)
    X_test_lda = lda_model.transform(X_test)
    results_lda = []
    for trainer in [train_knn, train_svm, train_mlp]:
        model, metrics, cm = trainer(X_train_lda, X_test_lda, y_train, y_test, "Com LDA")
        results_lda.append(metrics)
        plot_confusion_matrix(cm, metrics['model'], activity_labels)
    df_lda = pd.DataFrame(results_lda)
    print(df_lda[['model', 'accuracy', 'accuracy_std', 'precision', 'precision_std',
                  'recall', 'recall_std', 'f1_score', 'f1_score_std']].to_string(index=False))
    plot_comparison(df_lda, "Com LDA")

    return pd.concat([df_no_lda.assign(scenario='Sem LDA'),
                      df_lda.assign(scenario='Com LDA')], ignore_index=True)

# ============================================================================
# 8. EXECUÇÃO
# ============================================================================

if __name__ == "__main__":
    BASE_PATH = "./uci"
    final_results = run_experiments(BASE_PATH)
```
